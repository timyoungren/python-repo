```python
import os
import nltk as nltk
from nltk import metrics
from nltk.metrics import jaccard_distance
import pylab as pl
```

```python
jac_array = pl.random((40,40))
#jac_array = [[0.0 for j in range(40)] for i in range(40)]
dirstr = 'C:\\Users\\tim\\Documents\\U of C MScA\\2015 02 Big Data & Text Analytics\\Assignments\\Assignment 4\\Assignment Articles\\'
n = 4
outer_1 = 1
for root1, dirs1, files1 in os.walk(dirstr):
    for x in files1:
        inner_1 = 1
        file1 = open(str(os.path.join(root1,x)))
        raw1 = file1.read()
        gr1 = nltk.ngrams(raw1.split(), n)
        for root2, dirs2, files2 in os.walk(dirstr):
            for y in files2:
                file2 = open(str(os.path.join(root2,y)))
                raw2 = file2.read()
                gr2 = nltk.ngrams(raw2.split(), n)
                jac_array[outer_1-1][inner_1-1] = jaccard_distance(set(gr1),set(gr2))
                gr1 = nltk.ngrams(raw1.split(), n)
                inner_1 += 1
        outer_1 += 1
pl.pcolor(jac_array)
pl.colorbar()
pl.show()
```
