```python
f = open('C:\\Users\\Nick\\Documents\\Teaching\\University of Chicago\\Big Data Class\\Class 5\\Book\\3boat10.txt')
raw = f.read()

f2 = open('C:\\Users\\Nick\\Documents\\Teaching\\University of Chicago\\Big Data Class\\Class 5\\Book\\3boat10_out.txt', "w")


#Create your n-grams
n = 6
sixgrams = nltk.ngrams(raw.split(), n)

#compute frequency distribution for all the bigrams in the text
fdist = nltk.FreqDist(sixgrams)
for k,v in fdist.items():
    #print k,v
    f2.write(str(k)+"\t"+str(v)+"\n")
    
f2.close()
```
